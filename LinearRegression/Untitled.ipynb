```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
```

```python
df.head()
```

```python
plt.xlabel('year')
plt.ylabel('per_capita_income')
plt.scatter(df.year,df['per capita income (US$)'],color='blue',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2025]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
x = np.linspace(-10, 10, 100)
y = reg.coef_ * x + reg.intercept_
plt.plot(x, y, label=f'y = {reg.coef_}x + {reg.intercept_}')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
ls=[]
for y in df.year:
    # print(reg.predict([[y]]))
    ls.append(reg.predict([[y]]))
ls    
```

```python
import joblib as jb
```

```python
plt.scatter(df.year,df['per capita income (US$)'])
plt.plot(df.year,ls,color='red')
```

```python
df['new per capita income']=np.array(ls).ravel()
df.head()
```
